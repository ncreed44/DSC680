## Second Cleaning of Data

```python
college_players_df2 = college_players_df1.drop(college_players_df1.columns[[45, 44, 43, 42, 41, 40, 39, 38, 37, 36, 27]],  axis=1)
college_players_df2.head()
```

## Third Cleaning of Data

```python
college_players_df2['Rec Rank'] = college_players_df2['Rec Rank'].replace(np.nan,0) 
college_players_df2.head()
```

As mentioned in previous steps all player who did not have a Rec Rank, I replaced there Nan to a 0. Now there is no missing data.

## Fourth Cleaning of Data

```python
college_players_df3 = college_players_df2.drop_duplicates()
college_players_df3.head()
```

I confirmed there was on duplicates in the data set. There was no drop in rows so there were no duplicates.

## Fifth Cleaning of Data

```python
college_players_df3['ht'] = college_players_df3['ht'].replace('Jun-00','1-Jun') 
college_players_df3['ht'] = college_players_df3['ht'].replace(np.nan,'1-Jun')
college_players_df3.head() 
```

I changed all values that did not have values in the column ht to 1-Jun as well as all values that had a date that was impossible listed a Jun-00. This way if I want to see if the date variable is valuable, I can use it by combing it with year.
